<h2>1. Product Clicks </h2>

```python
df_clicks =  pd.read_csv("Tables/prod_Product_viewed.csv")
```

```python
df_clicks = df_clicks[['id', 'brand', 'category', 'event', 'event_text', 'name', 'original_timestamp', 'timestamp',\
                       'price', 'product_id', 'price', 'quantity', 'context_device_type']]
df_clicks
```

```python
df_clicks.info()
```

```python
brand_to_category_dict = dict(zip(df_clicks['brand'], df_clicks['category']))
brand_to_category_dict
```

```python
len(brand_to_category_dict.keys())
```

```python
df_clicks['clicks'] = np.zeros(len(df_clicks))
df_clicks = df_clicks.groupby('category', as_index=False).agg({'clicks': 'count'})
df_clicks.sort_values('clicks', ascending=False)
```

<h2> 2. Order Completed </h2>

<h2> 3. Adds to Wishlist </h2>

```python
df_wishlist = pd.read_csv('Tables/prod_adds_to_wishlist.csv')
df_wishlist = df_wishlist[['id', 'brand', 'category', 'event', 'original_timestamp', 'price', 'product_id']]
df_wishlist
```

```python
df_wishlist['no_of_adds_to_wishlist'] = np.zeros(len(df_wishlist))
df_wishlist = df_wishlist.groupby('category', as_index=False)[['category', 'no_of_adds_to_wishlist', 'price']].\
    agg({'price': 'sum', 'no_of_adds_to_wishlist': 'count'}).sort_values('no_of_adds_to_wishlist', ascending= False)
df_wishlist
```

<h2> 4. Adds to Cart </h2>

```python
df_cart = pd.read_csv('tables/prod_adds_to_cart.csv')
df_cart = df_cart[['brand', 'category', 'event', 'id', 'original_timestamp', 'price', 'product_id', 'quantity']]
df_cart
```

```python
df_cart['no_of_adds_to_cart'] = np.zeros(len(df_cart))
df_cart = df_cart.groupby('category', as_index=False)[['category', 'price',  'no_of_adds_to_cart']].agg({'price': 'sum', 'no_of_adds_to_cart': 'count'}).\
    sort_values('no_of_adds_to_cart', ascending=False)
df_cart
```
